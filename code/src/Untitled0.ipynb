```python
import requests
from bs4 import BeautifulSoup

def check_sanctioned_entity(entity_name):
    if entity_name=="":
      return False
    """
    Checks if an entity has search results on OpenSanctions.org.

    Args:
        entity_name (str): The name of the entity to search for.

    Returns:
        bool: True if search results are found, False otherwise.
    """
    entity_name = entity_name.replace(" ", "+")
    url = f"https://www.opensanctions.org/search/?scope=securities&q={entity_name}"
    try:
        response = requests.get(url)
        response.raise_for_status()  # Raise HTTPError for bad responses (4xx or 5xx)
        soup = BeautifulSoup(response.content, "html.parser")

        # Check for the presence of search results (adjust the selector as needed)
        # This is a basic example, you might need to inspect the page
        # structure to find the most robust selector.
        results = soup.find_all("div", class_="Search_resultTitle__twair")
        results2 = soup.find_all("span", class_="badge bg-warning")
        check_flag=False
        for result in results2:
          if result.text=="Sanctioned entity":
            check_flag= True
        return len(results) > 0 and check_flag

    except requests.exceptions.RequestException as e:
        print(f"Error fetching URL: {e}")
        return False

def set_sanctioned_entity_val(e1,e2,intermediary):
  if check_sanctioned_entity(e1):
    obj.sanct_entity_e1=1
  if check_sanctioned_entity(e2):
    obj.sanct_entity_e2=1
  if check_sanctioned_entity(intermediary):
    obj.sanct_entity_intermediary=1

print(check_sanctioned_entity("sberbank"))
```

```python
#normalisations
def fatf_norm(obj):
  present= (obj.curr_fatf_e1+obj.curr_fatf_e2+obj.curr_fatf_intermediary)/3
  past= (obj.hist_fatf_e1+obj.hist_fatf_e2+obj.hist_fatf_intermediary)/3
  return 0.7*present + 0.3*past
print("fatf score is ",fatf_norm(obj))
def corruption_norm(obj):
  return (obj.corruption_score_e1 + obj.corruption_score_e2 + obj.corruption_score_intermediary)/300
print("corruption score is ",corruption_norm(obj))
def basel_norm(obj):
  return (obj.basel_score_e1 + obj.basel_score_e2 + obj.basel_score_intermediary)*0.1/3
print("basel score is ",basel_norm(obj))



def country_norm(obj):
  return (fatf_norm(obj)+corruption_norm(obj)+basel_norm(obj))/3
print("country score is ",country_norm(obj))
```

```python
def sanctions_norm(persons,entities):
  if persons_norm(persons)!=0 and entities_norm(entities)!=0:
    return (persons_norm(persons)+entities_norm(entities))/2
  elif persons_norm(persons)!=0:
    return persons_norm(persons)
  elif entities_norm(entities)!=0:
    return entities_norm(entities)
  else:
    return 0

print("sanctions score is ",sanctions_norm(persons,entities))
def persons_norm(persons):
  var=0
  for person in persons:
    var=var+check_pep(person)
  return var/len(persons)
print("persons score is ",persons_norm(persons))
def entities_norm(entities):
  var=0
  for entity in entities:
    var=var+check_sanctioned_entity(entity)
  return var/len(entities)
print("entities score is ",entities_norm(entities))
```
